```python
import pandas as pd


df = pd.read_csv("Titanic-Dataset.csv")

print("Dataset Loaded Successfully!\n")


print("First 5 Rows:")
print(df.head(), "\n")

print("Dataset Info:")
print(df.info(), "\n")

print("Statistical Summary:")
print(df.describe(), "\n")


age_series = df['Age']
fare_series = df['Fare']
survived_series = df['Survived']

print("Missing Values Before Handling:")
print(df.isnull().sum(), "\n")

df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

print("Missing Values After Handling:")
print(df.isnull().sum(), "\n")


print("Using loc:")
print(df.loc[0:5, ['Name', 'Age', 'Survived']], "\n")

print("Using iloc:")
print(df.iloc[0:5, 1:6], "\n")


df.rename(columns={'Sex': 'Gender'}, inplace=True)


print("Passengers Age > 30:")
print(df[df['Age'] > 30].head(), "\n")

print("Passengers Fare > 50:")
print(df[df['Fare'] > 50].head(), "\n")

print("Female Passengers Who Survived:")
print(df[(df['Gender'] == 'female') & (df['Survived'] == 1)].head(), "\n")


print("Sorted by Fare (Descending):")
print(df.sort_values(by='Fare', ascending=False).head(), "\n")

print("Survival Rate by Gender:")
print(df.groupby('Gender')['Survived'].mean(), "\n")

print("Average Age by Passenger Class:")
print(df.groupby('Pclass')['Age'].mean(), "\n")

print("Survival Count by Embarked Location:")
print(df.groupby('Embarked')['Survived'].sum(), "\n")


pivot_table = df.pivot_table(
    values='Survived',
    index='Gender',
    columns='Pclass',
    aggfunc='mean'
)

print("Pivot Table (Survival Rate):")
print(pivot_table, "\n")


class_info = pd.DataFrame({
    'Pclass': [1, 2, 3],
    'ClassType': ['Upper', 'Middle', 'Lower']
})

merged_df = pd.merge(df, class_info, on='Pclass')

print("Merged DataFrame Sample:")
print(merged_df.head(), "\n")


df_part1 = df[:400]
df_part2 = df[400:]

combined_df = pd.concat([df_part1, df_part2])


avg_age = df.groupby('Survived')['Age'].mean()
max_fare = df['Fare'].max()
min_fare = df['Fare'].min()
survival_percentage = (df['Survived'].sum() / len(df)) * 100

print("Average Age (0 = Died, 1 = Survived):")
print(avg_age, "\n")

print("Highest Fare Paid:", max_fare)
print("Lowest Fare Paid:", min_fare)

print(f"Overall Survival Percentage: {survival_percentage:.2f}%")

print("\n--- ANALYSIS COMPLETE ---")
```
